```python
import numpy as np
from sklearn.metrics import accuracy_score
from google.colab import drive
import time
drive.mount('/content/drive')
```

```python
# Load the data
LOAD_TRAIN = np.load('/content/drive/MyDrive/train.npz')
LOAD_TEST = np.load('/content/drive/MyDrive/test.npz')
x1 = LOAD_TRAIN['inputs']
y1 = LOAD_TRAIN['outputs']
x2 = LOAD_TEST['inputs']
y2 = LOAD_TEST['outputs']
random_states = [40, 42, 69, 72, 99] 
```

```python
# Helper function to print statistics related to training time and accuracy

def time_and_accuracy(time_1, accuracy_1):
    print('Average time =', sum(time_1)/len(time_1))
    print('Average accuracy =', sum(accuracy_1)/len(accuracy_1))
```

# Problem 1

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

accuracy_decision = list()
time_decision = list()

for i in random_states:
    start = time.time()
    tree_clf = DecisionTreeClassifier(max_depth=4, random_state=i)
    tree_clf.fit(x1,y1)
    t = time.time() - start
    time_decision.append(t)
    print('Training time =', t, 'seconds')
    y_pred_tree = tree_clf.predict(x2)
    print('Accuracy for Decision Tree =', accuracy_score(y2, y_pred_tree))
    accuracy_decision.append(accuracy_score(y2, y_pred_tree))
    print()
time_and_accuracy(time_decision, accuracy_decision)
```

```python
# Bagging Decision Tree
from sklearn.ensemble import BaggingClassifier
accuracy_bagging = list()
time_bagging = list()

for i in random_states:
    start = time.time()
    bag_clf = BaggingClassifier(
        DecisionTreeClassifier(splitter = 'random', max_leaf_nodes = 16,random_state=i), n_estimators=500,
        max_samples=500, max_features=20, bootstrap=True, n_jobs=-1, random_state=i)
    bag_clf.fit(x1, y1)
    t = time.time() - start
    time_bagging.append(t)
    print('Training time =', t, 'seconds')
    y_pred_bagging = bag_clf.predict(x2)
    print('Accuracy for Bagging Decision Tree =', accuracy_score(y2, y_pred_bagging))
    accuracy_bagging.append(accuracy_score(y2, y_pred_bagging))
    print()

time_and_accuracy(time_bagging, accuracy_bagging)
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression

accuracy_log = list()
time_log = list()

for i in random_states:
    start = time.time()
    log_reg = LogisticRegression(multi_class ='multinomial', solver='lbfgs', C=10, random_state = i)
    log_reg.fit(x1, y1)
    t = time.time()-start
    time_log.append(t)
    print('Training time =', t, 'seconds')
    y_pred_logreg = log_reg.predict(x2)
    print('Accuracy for Logistic Regression =', accuracy_score(y2, y_pred_logreg))
    accuracy_log.append(accuracy_score(y2, y_pred_logreg))
    print()

time_and_accuracy(time_log, accuracy_log)
```

```python
# Linear Support Vector Machine
from sklearn import svm

accuracy_svm = list()
time_svm = list()
for i in random_states:
    start = time.time()
    linear_svm = svm.LinearSVC(C=1, loss='hinge', max_iter=1000)
    linear_svm.fit(x1, y1)
    t = time.time()-start
    print('Training time =', t, 'seconds')
    time_svm.append(t)
    y_pred_svm = linear_svm.predict(x2)
    print('Accuracy for Linear SVM =', accuracy_score(y2, y_pred_svm))
    accuracy_svm.append(accuracy_score(y2, y_pred_svm))
    print()

time_and_accuracy(time_svm, accuracy_svm)
```

# Problem 3

```python
import torch
from torch import nn
from torch.utils.data import TensorDataset
from torch.utils.data import DataLoader
from torch import Tensor

class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.layer1 = nn.Sequential(
          nn.Conv1d(1, 4, kernel_size = 4, stride = 4),
          nn.ReLU(),
          torch.nn.AvgPool1d(kernel_size = 4, stride = 4)
        )
        self.layer2 = nn.Sequential(
          nn.Conv1d(4, 4, kernel_size = 3, stride = 3, padding = 1),
          nn.ReLU(),
          nn.AvgPool1d(kernel_size = 3, stride = 3)
        )
        self.fc = nn.Linear(4*3, 3, bias = True)
    
    def forward(self, input):
        output = self.layer1(input)
        output = self.layer2(output)
        output = output.view(output.size(0), -1) # flatten
        output = self.fc(output)
        return output
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
loss_ft = nn.CrossEntropyLoss().to(device)

trainx_tensor = Tensor(x1, device = device) # Convert training inputs to tensor
trainy_tensor = Tensor(y1, device = device) # Convert training outputs to tensor
testx_tensor = Tensor(x2, device = device)  # Convert test inputs to tensor
testy_tensor = Tensor(y2, device = device)  # Convert test outputs to tensor
```

```python
accuracy_cnn = list() # temporary value to store average accuracy
time_cnn = list()

for i in range(5):
    start = time.time()
    model = CNN().to(device)
    optimizer = torch.optim.Adam(model.parameters(), lr = 0.005)
    data_loader = DataLoader(dataset = TensorDataset(trainx_tensor, trainy_tensor),
                                          batch_size = 30, shuffle = True)
    model.train()
    for j in range(20):
        for x, y in data_loader:
            x = x.to(device)[:, None, :]
            y = y.to(device).long()
            
            optimizer.zero_grad()
            pred = model(x.float())
            loss = loss_ft(pred, y)
            loss.backward()
            optimizer.step()
    t = time.time()-start
    print('Time for CNN =', t)
    time_cnn.append(t)
    model.eval()

    with torch.no_grad():
        prediction = model(testx_tensor[:, None, :])
        correct_prediction = torch.argmax(prediction, 1) == testy_tensor
        accuracy = correct_prediction.float().mean()
        accuracy_cnn.append(accuracy.item())
        print('Accuracy for CNN =', accuracy.item())
        print()
time_and_accuracy(time_cnn, accuracy_cnn)
```
